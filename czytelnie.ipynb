**Zadanie 2**

**Które z następujących problemów można uznać za mieszczące się w zakresie sztucznej
inteligencji:**

* Streszczanie tekstu – **Tak. Modele AI potrafią skracać teksty na podstawie analizy treści.**

* Tłumaczenie tekstu – **Tak. AI, jak Google Translate, skutecznie tłumaczy teksty.**

* Klasyfikacja tekstu do kategorii tematycznych – **Tak. AI może klasyfikować teksty za pomocą algorytmów przetwarzania języka naturalnego.**

* Odpowiadanie na proste pytania zadawane w języku naturalnym – **Tak. Chatboty i modele językowe mogą odpowiadać na proste pytania.**

* Rozwiązywanie układów równań – **Tak. AI może rozwiązywać układy równań przy użyciu algorytmów matematycznych.**

* Układanie rozkładu jazdy – **Tak. AI może optymalizować rozkłady jazdy, biorąc pod uwagę wiele zmiennych.**

* Rozwiązywanie układów równań liniowych – **Tak. To zadanie, które AI może wykonywać szybko i efektywnie.**

* Symboliczne obliczanie pochodnych – **Tak. AI może obliczać pochodne za pomocą systemów algebraicznych.**

* Symboliczne całkowanie – **Tak. AI, jak WolframAlpha, potrafi przeprowadzać symboliczne całkowanie.**

* Kierowanie samochodem – **Tak. Autonomiczne pojazdy są oparte na zaawansowanej AI do analizy otoczenia i podejmowania decyzji.**

obecnie skutecznie imitowane przez sztuczne systemy (odpowiednio oprogramowane

* Odpowiadanie na pytania klientów w telefonicznej infolinii – **Częściowo. AI może obsługiwać proste pytania, ale bardziej złożone sytuacje wymagają ludzkiego wsparcia.**
* Odpowiadanie na pytania klientów w internetowej infolinii – **Tak. Chatboty AI dobrze radzą sobie z typowymi zapytaniami klientów online.**
